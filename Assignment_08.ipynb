```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HousingData.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['CRIM'] = df['CRIM'].fillna(df['CRIM'].median())
df['ZN'] = df['ZN'].fillna(df['ZN'].median())
df['CHAS'] = df['CHAS'].fillna(df['CHAS'].median())
df['INDUS'] = df['INDUS'].fillna(df['INDUS'].median())
df['LSTAT'] = df['LSTAT'].fillna(df['LSTAT'].median())
df['AGE'] = df['AGE'].fillna(df['AGE'].median())
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
sns.boxplot(data = df)
```

```python
df.skew()
```

```python
def outlier_removal(df, attribute):
    q1 = df[attribute].quantile(0.25)
    q3 = df[attribute].quantile(0.75)
    iqr = q3 - q1
    df = df[(df[attribute] > q1 - 1.5*iqr) & (df[attribute] < q3 + 1.5*iqr)]
```

```python
outlier_removal(df, 'CRIM')
outlier_removal(df, 'CHAS')
```

```python
df.skew()
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler_df = scaler.fit_transform(X_train)
```

```python
scaler_df
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
regression = LinearRegression()
regression.fit(scaler_df, Y_train)
```

```python
cross_val_score(regression, scaler_df, Y_train, scoring = "neg_mean_squared_error", cv = 5)
```

```python
reg_predict = regression.predict(X_test)
```

```python
reg_predict
```

```python
sns.displot(reg_predict - Y_test, kind = 'kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_predict, Y_test)
```
