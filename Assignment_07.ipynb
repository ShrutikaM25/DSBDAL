```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
doc1 = open("doc1", "r").read()
```

```python
import nltk
from nltk import word_tokenize, sent_tokenize, StanfordTagger, corpus, stem
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

# Tokenization

```python
text = word_tokenize(doc1)
print(text)
```

```python
sentence = sent_tokenize(doc1)
print(sentence)
```

# POS Tagging

```python
tag = nltk.pos_tag(text)
print(tag)
```

# Stopword Removal

```python
stop_remove = stopwords.words('english')
print(stop_remove)
```

```python
filtered_text = ""

for word in text :
  if word not in stop_remove:
    filtered_text+= word + " "

print(filtered_text)
```

# Stemming

```python
ps = PorterStemmer()

for word in text :
  print (word , " : ", ps.stem(word) )
```

# Lemmatization

```python
lemmatizer = WordNetLemmatizer()

for word in text:
  print(word, " : ", lemmatizer.lemmatize(word))

  # print(word, " : ", lemmatizer.lemmatize(word, pos="a"))
```

```python
def term_freq (doc):
    tokens = word_tokenize(doc)
    dic = {}
    for word in tokens:
      if word in dic:
        dic[word]+=1
      else:
        dic[word] = 1

    for word, freq in dic.items():
        dic[word] = freq/len(tokens)
        print(word, " : ", dic[word])

term_freq(doc1)
```

```python
import math
def idf():
  doc1_token = word_tokenize(doc1)
  doc2 = open("doc2", "r").read()
  doc2_token = word_tokenize(doc2)
  dic = {}
  doc_count = 2
  all_tokens = doc1_token + doc2_token
  for word in all_tokens:
    dic[word] = 1
    if word in doc1_token:
      dic[word] += 1
    if word in doc2_token:
      dic[word] +=1
  for word, freq in dic.items():
    dic[word] = math.log(doc_count/freq)
    print(word, " : ", dic[word])

idf()
```
