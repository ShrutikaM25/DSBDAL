```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.describe()
```

Distribution of Age with respect to each gender along with survival information

```python
df['Age_range'] = pd.qcut(df['Age'], 5)
```

```python
sns.catplot(x = 'Sex', col='Survived' ,hue = 'Age_range', data = df, kind = 'count')
```

```python
sns.boxplot(x= 'Sex', y= 'Age', hue = 'Survived', data = df )
```
