# Z-Score Normalization

```python
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(df['math score']))
print(z)
```

```python
z.hist()
```

```python
sns.displot(data=df['math score'],bins=25,kind='hist',aspect=1.5,kde=True)
```

# Normalization

```python
q1 = df['math score'].quantile(0.25)
q3 = df['math score'].quantile(0.75)

iqr = q3 - q1

lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

df = df[(df['math score'] > lower) & (df['math score'] < upper)]
df.shape
```

```python
sns.boxplot(df['math score'])
```

# Data Transformation

```python
# Data Transformation
sns.displot(data = df['math score'], kde = True) # Almost normal
df['math score'].skew()
```
